```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow 
```

```python
from google.colab import drive
drive.mount('/content/drive')

img_path = "/content/drive/MyDrive/CodeClause/Singapore.jpg"
img = cv2.imread(img_path)
img2 = cv2.resize(img, (0, 0), fx=0.5, fy=0.5)
```

```python
# Function to convert image to pencil sketch
def convert_to_sketch(img):
    gray_image = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
    
   
    inverted_image = 255 - gray_image
    
 
    blurred_image = cv2.GaussianBlur(inverted_image, (21,21), 0)
    
   
    inverted_blurred = 255 - blurred_image
    
    
    pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
    
    return pencil_sketch
```

```python
# Convert the image to pencil sketch
sketch_img = convert_to_sketch(img2)


cv2_imshow(img2)
print(" ")
cv2_imshow(sketch_img)
cv2.waitKey(0)
```
